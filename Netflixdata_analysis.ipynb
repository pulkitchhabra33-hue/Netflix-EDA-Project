```python
df.drop_duplicates(inplace= True)
df
```

```python
df['title']= df['title'].str.strip()
```

```python
df['date_added']= df['date_added'].astype(str).str.strip()
df['date_added']= pd.to_datetime(df['date_added'], format= "%B %d, %Y")
df
```

```python
df['year_added']= df['date_added'].dt.year
```

```python
df['year_added']= df['year_added'].astype('Int64')
```

```python
df['month_name']= df['date_added'].dt.month_name()
```

```python
year_counts= df['year_added'].value_counts().sort_index()
year_counts
```

```python
sns.countplot(x= 'rating', data= df, order= df['rating'].value_counts().index)
plt.xticks(rotation= 90)
plt.title('Distribution of Rating')
plt.show()
```

```python
sns.barplot(x='type', y= 'year_added', data= df)
plt.show()
```

```python
type_year= df.groupby(['year_added', 'type']).size().reset_index(name= 'count')
type_year
```

```python
sns.barplot(x= 'year_added', y= 'count', hue= 'type', data= type_year)
plt.title('Movies vs TV Shows added per year')
plt.xticks(rotation= 90)
plt.show()
```

```python
country_count= df.groupby(['country', 'type']).size().reset_index(name= 'count_country').head(10)
country_count.index= range(1, len(country_count)+1)
country_count
```

```python
sns.barplot(x= 'country', y= 'count_country', hue='type', data= country_count)
plt.title('Movie vs TV Shows of top countries')
plt.xticks(rotation= 90)
plt.show()
```

```python
rating_count= df.groupby(['rating', 'type']).size().reset_index(name= 'count_rating')
# country_count.index= range(1, len(country_count)+1)
rating_count
```

```python

sns.barplot(x= 'rating', y= 'count_rating', hue='type', data= rating_count)
plt.title('Movie vs TV Shows on Rating Basis')
plt.xticks(rotation= 90)
plt.show()
```

```python
#lineplot for Content added per year 
sns.lineplot(x= year_counts.index, y= year_counts.values)
plt.title('Content added per year')
plt.xlabel('Year')
plt.ylabel('Numbner of titles')
plt.xticks(rotation= 90)
plt.show()
```

```python
#barplot for Content added per year 
sns.barplot(x= year_counts.index, y= year_counts.values)
plt.title('Content added per year')
plt.xlabel('Year')
plt.ylabel('Numbner of titles')
plt.xticks(rotation= 90)
plt.show()
```

```python
#piechart for Ratings distribution

rating_counts= df['rating'].value_counts().head(10)
rating_counts.plot.pie(autopct= '%1.1f%%', figsize= (10,10))
plt.title('Ratings Distribution')
plt.ylabel("")
plt.show()
```

```python
top_directors= df['director'].value_counts().head(10).index
df_directors= df[df['director'].isin(top_directors)]
```

```python
sns.countplot(x= 'director', data= df_directors, order= top_directors)
plt.title('Top 10 Directors')
plt.xticks(rotation= 90)
plt.show()
```

```python
movies= df[df['type']== 'Movie'].copy()
movies['duration']= movies['duration'].fillna("0 min")
movies['duration_minutes']= movies['duration'].str.replace(" min","").astype(int)
```

```python
sns.kdeplot(x= movies['duration_minutes'], fill= True)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (Minutes)')
plt.show()
```

```python
print("📌 NETFLIX EDA – CONCLUSION SUMMARY\n")

print("1. Movies vs TV Shows:")
print(f"   - Movies: {df['type'].value_counts().get('Movie', 0)}")
print(f"   - TV Shows: {df['type'].value_counts().get('TV Show', 0)}\n")

print("2. Content Added Per Year (Trend):")
print(f"   - Earliest year: {df['year_added'].min()}")
print(f"   - Latest year: {df['year_added'].max()}")
print("   - Content has increased steadily over the years.\n")

print("3. Top 5 Countries Producing Most Content:")
print(df['main_country'].value_counts().head(5), "\n")

print("4. Ratings Distribution (Top Ratings):")
print(df['rating'].value_counts().head(5), "\n")

print("5. Top 5 Genres:")
print(df['main_genre'].value_counts().head(5), "\n")

print("6. Average Movie Duration:")
if 'duration_minutes' in movies.columns:
    print(f"   - Average Duration: {movies['duration_minutes'].mean():.2f} minutes\n")
else:
    print("   - Duration data not available.\n")

print("7. Missing Values Summary:")
print(df.isnull().sum(), "\n")

print("✨ Overall Insight:")
print("   Netflix has more Movies than TV Shows, content production grew rapidly in the last decade,")
print("   the United States and India dominate content creation, TV-MA and TV-14 are the most common ratings,")
print("   Drama and Comedy are top genres, and movies usually last around 90–110 minutes.")

```
